```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://code.datasciencedojo.com/datasciencedojo/datasets/raw/master/Daily%20Demand%20Forecasting%20Orders/Daily_Demand_Forecasting_Orders.csv",sep=";",names=["week_of_month","day_of_week","non-urgent_order","urgent_order","A_type","B_type","C_type","fiscal_orders","traffic_controller","1_banking","2_banking","3_banking","target"])
df = df.iloc[1:,:]
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.astype(float)
```

```python
df.head()
```

```python
df.week_of_month = pd.Categorical(df.week_of_month)
df.day_of_week = pd.Categorical(df.day_of_week)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sscaler = StandardScaler()
```

```python
cols_standard = ["non-urgent_order","urgent_order","A_type","B_type","C_type","fiscal_orders","traffic_controller","1_banking","2_banking","3_banking"]
```

```python
df[cols_standard] = sscaler.fit_transform(df[cols_standard])
```

```python
df.drop_duplicates()
```

```python
df.head()
```

```python
X = df.drop(columns="target")
y = df.target
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15)
```

```python
list_of_model =[DecisionTreeRegressor(), KNeighborsRegressor(), MLPRegressor(), GradientBoostingRegressor(), RandomForestRegressor(), SVR(), LGBMRegressor()]
```

```python
def model_compare():
    score_list=list()
    for model in list_of_model:
        cv_score = np.mean(cross_val_score(model,X,y,cv=6))
        estimator = model.fit(X_train,y_train)
        y_pred = estimator.predict(X_test)
        mse_score = mean_squared_error(y_test,y_pred)
        score_list.append({
            "model":model,
            "mean_squared_error" : mse_score,
            "r_mean_squared_error" : mse_score**0.5,
            "cross_val_score" : cv_score,
            "r2_score" : r2_score(y_test,y_pred)
        })
    result_df = pd.DataFrame(score_list,columns=["model","mean_squared_error","r_mean_squared_error","cross_val_score","r2_score"])
    return result_df
```

```python
result_df = model_compare()
```
